```
#Importing both Pandas and Numpy Library
import pandas as pd
import numpy as np
```

```
#Loading and viewing the CDR_descriptiont data set for the project
df1 = pd.read_excel('CDR_description.xlsx')

# Lastly, view the first ten rows
df1.head()
```

```
#Loading and viewing the CDR_descriptiont data set for the project
df2 = pd.read_excel('cells_geo_description.xlsx')
df2.head()
```

```
#Loading Telcom_dataset1 
df3 = pd.read_csv('Telcom_dataset.csv') 
df3
```

```
#Loading Telcom_dataset2 
df4 = pd.read_csv('Telcom_dataset2.csv') 
df4
```

```
#Loading Telcom_dataset3
df5 = pd.read_csv('Telcom_dataset3.csv') 
df5
```

```
#Loading cells geo dataset
df6 = pd.read_csv('cells_geo.csv', delimiter=';') 
df6
```

```
#Changing column names of site ID and Cell ID of dataset 3

df5.rename(columns={'CELLID':'CELL_ID',
                   'SIET_ID':'SITE_ID'},inplace='True')
df5
```

```
#Changing column names of site ID and Cell ID of dataset 2

df4.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT',
                   'DW_B_NUMBER':'DW_B_NUMBER_INT'},inplace='True')
df4.head()
```

```
#Changing column names of site ID and Cell ID of dataset 3

df3.rename(columns={'PRODUTC':'PRODUCT',
                     'DATETIME':'DATE_TIME'},inplace='True')
df3.head()
```

```
#Combining all my three telcom datasets
telcom = pd.concat([df3, df4, df5], axis=0)
telcom
```

```
#Cleaning my Telcom dataset
#First am checking for any null values
telcom.isnull().any()
```

```
#Check what exactly is missing on site id column
telcom.loc[telcom['SITE_ID'].isnull(),'SITE_ID']

#Deleting the rows with no values
telcom.dropna(inplace=True)
telcom.head()
```

```
#Checking for duplicate values on telcom merged dataset and later dropping them
telcom.duplicated()

#I find there are duplicate values and drop them
telcom.drop_duplicates()

#Checking on duplicate columns
telcom.columns.duplicated()

#The below dataset shows there are no duplicate columns
```

```
#The below code joins telcom datasets and Geo dataset

df = pd.merge(telcom, df6,how = 'inner', left_on='SITE_ID', right_on='SITE_CODE')
df
```

```
#Checking for any duplicate values and later dropping them
df.isnull().any()

#I find there are missing values on Status and area columns and Drop the null values
df.dropna(inplace=True)
df.head()
```

```
#Checking for any duplicates on my dataset and drop any duplicates
df.duplicated()

#I find there are duplicate values and drop them
df.drop_duplicates()

#Checking for duplicate columns 
df.columns.duplicated()

#From below output there are no duplicate columns
```

```
# The below code finds the most used city in the three 

df.VILLES.value_counts().head(10)
```

```
#To find out which product is the most used
df.PRODUCT.value_counts()
```

```
#Grouping both sms and voice by value to find the different
df.groupby(["PRODUCT"])["VALUE"].mean()
```

```
#To find out which localisation has most Cities
df.LOCALISATION.value_counts()
```

```
#Finding the trend in cell on site 
df.groupby(["PRODUCT"])["CELL_ON_SITE"].value_counts()
```

```
#To find the status of all cities in the dataset to see in all active

df["STATUS"].unique()     #All status show in service
```

```
#To find out which product was most used in both home and business hours
mostUsed = df.groupby(by='PRODUCT').agg('count').reset_index()
mostUsed
```
